```python
from PIL import Image
import numpy as np
import cv2
import os

def generate_dataset():
    face_classifier = cv2.CascadeClassifier('haar_face.xml')
    
    def cropped_face(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        if len(faces) == 0:
            return None
        for (x, y, w, h) in faces:
            face_crop = gray[y:y+h, x:x+w]
            return face_crop

    cap = cv2.VideoCapture(0)
    img_id = 0
    id = 1
    

    

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break
        
        face = cropped_face(frame)
        if face is not None:
            img_id += 1
            face_resized = cv2.resize(face, (200, 200))
            
            path = r"C:\face detaction prototype\images\image."+str(id)+"."+str(img_id)+".jpg"
            cv2.imwrite(path, face_resized)
            
            
            if os.path.exists(path):
                print(f"Image saved at: {path}")
            else:
                print(f"Failed to save image at: {path}")

            cv2.putText(face_resized, str(img_id), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
            cv2.imshow("Cropped image", face_resized)

        if cv2.waitKey(1) == 13 or img_id == 500:
            break

    cap.release()
    cv2.destroyAllWindows()
    print("Done")
```

```python
generate_dataset()
```

```python
import os
def train_classifier(data):
    path=[os.path.join(data,f) for f in os.listdir(data)]
    faces=[]
    id=[]
    for image in path:
        img=Image.open(image).convert('L')
        imageNp=np.array(img,'uint8')
        faces.append(imageNp)
        ids=int(os.path.split(image)[1].split(".")[1])
        id.append(ids)
        
    id=np.array(id)
    print(id)
    clf = cv2.face.LBPHFaceRecognizer_create()

    clf.train(faces,id)
    clf.write("classifier.xml")
```

```python
train_classifier("images")
```

```python
from datetime import date
import cv2
import csv
file_name="attendace.csv"
recorded_names = set()

header = ["Name", "Time"]
with open(file_name, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(header)
def draw_boundary(img, classifier, scaleFactor, minNeighbors, color, text, clf):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    features = classifier.detectMultiScale(gray_img, scaleFactor, minNeighbors)

    for (x, y, w, h) in features:
        cv2.rectangle(img, (x, y), (x+w, y+h), color, 2)
        id, confidence = clf.predict(gray_img[y:y+h, x:x+w])
        print(confidence)
        if confidence>35:
            if id==0:
                cv2.putText(img, "Prakash", (x,y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
                name = "Prakash"
                today = date.today().strftime("%Y-%m-%d")

            elif id==1:
                cv2.putText(img, "Abhisek", (x,y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
                name = "Abhisek"
                today = date.today().strftime("%Y-%m-%d")

            if name not in recorded_names and name != "UNKNOWN":
                with open(file_name, mode="a", newline="") as file:
                    writer = csv.writer(file)
                    writer.writerow([name, today])
                recorded_names.add(name)
            
            
            
            
            
        else:
            cv2.putText(img, "UNKNOWN", (x,y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,0,255), 1, cv2.LINE_AA)
        
    
    return img

video_capture = cv2.VideoCapture(0)
faceCascade = cv2.CascadeClassifier('haar_face.xml')
clf = cv2.face.LBPHFaceRecognizer_create()
clf.read("classifier.xml")



while True:
    ret, img = video_capture.read()

    if not ret:
        print("Failed to capture image")
        continue

    img = draw_boundary(img, faceCascade, 1.3, 3, (255, 255, 255), "Abhishek", clf)
    cv2.imshow("Face Detection", img)

    if cv2.waitKey(1) == 13:  
        break

video_capture.release()
cv2.destroyAllWindows()
```
